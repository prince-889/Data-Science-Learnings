Scatter Plot (sns.scatterplot)

1. Use: A scatter plot displays the relationship between two numerical variables. Each point represents an observation in the dataset.

Example: Plotting height vs. weight to observe the correlation between them.
Box Plot (sns.boxplot)

2. Use: A box plot (or box-and-whisker plot) provides a summary of a set of data values, showing the median, quartiles, and outliers. It's useful for comparing distributions.

Example: Comparing test scores across different groups.
Histogram (sns.histplot)

3. Use: A histogram displays the distribution of a single numerical variable by dividing the data into bins and counting the number of observations in each bin.

Example: Showing the distribution of ages in a population.
Heatmap (sns.heatmap)

4. Use: A heatmap is a graphical representation of data where individual values are represented as colors. It's often used to display correlation matrices or missing data patterns.

Example: Visualizing the correlation between different features in a dataset.
Line Plot (sns.lineplot)

5. Use: A line plot is used to display data points along a continuous variable, often used to show trends over time.

Example: Plotting the stock prices over a period of time to observe trends and patterns.

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")


flights_pivot = flights.pivot(index="month", columns="year", values="passengers")


plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Monthly Passengers Over the Years')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
